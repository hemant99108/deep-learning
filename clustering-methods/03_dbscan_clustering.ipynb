# dbscan clustering helps to do clusterin with non-linear data

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons  #create two half moons structure type data 
import matplotlib.pyplot as plt 
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
X,y=make_moons(n_samples=250,noise=0.05)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
#standard scaling 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
#dbscan algorithm 
dbscan=DBSCAN(eps=0.5)
```

```python
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```
